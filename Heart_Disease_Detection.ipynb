```python
df = data[(data['RestingBP']>70) & (data['RestingBP']<170)]
```

```python
df
```

```python
df = df[(df['Cholesterol']>80) & (df['Cholesterol']<400)]
```

```python
df = df[(df['MaxHR']>70)]
```

```python
df = df[(df['Oldpeak']>-2) & (df['Oldpeak']<3.7)]
```

```python
df
```

```python
X = df.loc[:,:'ST_Slope']
y = df['HeartDisease'] 
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python

model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test) 
```

```python
pred
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix
ac = accuracy_score(y_test,pred)
ac
```

```python
cf = confusion_matrix(y_test,pred)
sns.heatmap(cf,annot=True)
```

```python
sns.scatterplot(y_test,pred)
```

**Using KNN Algorithm**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df1 = df.drop(['HeartDisease'],axis=1)
df1 = pd.get_dummies(df1,drop_first=True)
scaler.fit(df1)
```

```python
scaler_feature = pd.DataFrame(scaler.transform(df1),columns=df1.columns)
scaler_feature
```

```python
sns.pairplot(df,hue="HeartDisease")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaler_feature,df['HeartDisease'],test_size=0.15,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
pred = model.predict(x_test)
print(classification_report(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
accuracy_rate = []
for i in range(1,40):
  model = KNeighborsClassifier(n_neighbors=i)
  score = cross_val_score(model,scaler_feature,df['HeartDisease'],cv=10)
  accuracy_rate.append(score.mean())
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),accuracy_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
